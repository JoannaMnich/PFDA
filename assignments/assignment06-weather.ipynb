## Temperture

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = "hly4935.csv"

# Read CSV
df = pd.read_csv(file_path, header=20, low_memory=False)

# Convert 'date' to datetime
df['date'] = pd.to_datetime(df['date'], errors='coerce')

# Drop rows with missing date
df = df.dropna(subset=['date'])

# Set date as index
df.set_index('date', inplace=True)

# Convert 'temp' to numeric and drop NaNs
df['temp'] = pd.to_numeric(df['temp'], errors='coerce')
df = df.dropna(subset=['temp'])

# Plot temperature
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['temp'], color='orange')
plt.xlabel('Date')
plt.ylabel('Temperature (C)')
plt.title('Air Temperature Over Time')
plt.grid(True)
plt.show()

```

## The Mean temperature each day

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = "hly4935.csv"

# Read CSV
df = pd.read_csv(file_path, header=20, low_memory=False)

# Convert 'date' to datetime and drop missing
df['date'] = pd.to_datetime(df['date'], errors='coerce')
df = df.dropna(subset=['date'])

# Set date as index
df.set_index('date', inplace=True)

# Convert 'temp' to numeric and drop NaNs
df['temp'] = pd.to_numeric(df['temp'], errors='coerce')
df = df.dropna(subset=['temp'])

# Calculate daily mean temperature
daily_mean = df['temp'].resample('D').mean()

# Plot daily mean temperature
plt.figure(figsize=(12, 6))
plt.plot(daily_mean.index, daily_mean, color='blue', label='Daily Mean Temp')
plt.xlabel('Date')
plt.ylabel('Temperature (C)')
plt.title('Daily Mean Air Temperature')
plt.legend()
plt.grid(True)
plt.show()
```

## The mean temperature for each month

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = "hly4935.csv"

# Read CSV
df = pd.read_csv(file_path, header=20, low_memory=False)

# Convert 'date' to datetime and drop missing
df['date'] = pd.to_datetime(df['date'], errors='coerce')
df = df.dropna(subset=['date'])

# Set date as index
df.set_index('date', inplace=True)

# Convert 'temp' to numeric and drop NaNs
df['temp'] = pd.to_numeric(df['temp'], errors='coerce')
df = df.dropna(subset=['temp'])

# Calculate monthly mean temperature
monthly_mean = df['temp'].resample('M').mean()

# Plot monthly mean temperature
plt.figure(figsize=(12, 6))
plt.plot(monthly_mean.index, monthly_mean, color='red', label='Monthly Mean Temp')
plt.xlabel('Date')
plt.ylabel('Temperature (C)')
plt.title('Monthly Mean Air Temperature')
plt.legend()
plt.grid(True)
plt.show()
```
