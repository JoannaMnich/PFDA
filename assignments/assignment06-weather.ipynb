## Temperture

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = "hly4935.csv"

# Read CSV
df = pd.read_csv(file_path, header=20, low_memory=False)

# Convert 'date' to datetime
df['date'] = pd.to_datetime(df['date'], errors='coerce')

# Drop rows with missing date
df = df.dropna(subset=['date'])

# Set date as index
df.set_index('date', inplace=True)

# Convert 'temp' to numeric and drop NaNs
df['temp'] = pd.to_numeric(df['temp'], errors='coerce')
df = df.dropna(subset=['temp'])

# Plot temperature
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['temp'], color='orange')
plt.xlabel('Date')
plt.ylabel('Temperature (C)')
plt.title('Air Temperature Over Time')
plt.grid(True)
plt.show()

```

## The Mean temperature each day

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = "hly4935.csv"

# Read CSV
df = pd.read_csv(file_path, header=20, low_memory=False)

# Convert 'date' to datetime and drop missing
df['date'] = pd.to_datetime(df['date'], errors='coerce')
df = df.dropna(subset=['date'])

# Set date as index
df.set_index('date', inplace=True)

# Convert 'temp' to numeric and drop NaNs
df['temp'] = pd.to_numeric(df['temp'], errors='coerce')
df = df.dropna(subset=['temp'])

# Calculate daily mean temperature
daily_mean = df['temp'].resample('D').mean()

# Plot daily mean temperature
plt.figure(figsize=(12, 6))
plt.plot(daily_mean.index, daily_mean, color='blue', label='Daily Mean Temp')
plt.xlabel('Date')
plt.ylabel('Temperature (C)')
plt.title('Daily Mean Air Temperature')
plt.legend()
plt.grid(True)
plt.show()
```

## The mean temperature for each month

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = "hly4935.csv"

# Read CSV
df = pd.read_csv(file_path, header=20, low_memory=False)

# Convert 'date' to datetime and drop missing
df['date'] = pd.to_datetime(df['date'], errors='coerce')
df = df.dropna(subset=['date'])

# Set date as index
df.set_index('date', inplace=True)

# Convert 'temp' to numeric and drop NaNs
df['temp'] = pd.to_numeric(df['temp'], errors='coerce')
df = df.dropna(subset=['temp'])

# Calculate monthly mean temperature
monthly_mean = df['temp'].resample('M').mean()

# Plot monthly mean temperature
plt.figure(figsize=(12, 6))
plt.plot(monthly_mean.index, monthly_mean, color='red', label='Monthly Mean Temp')
plt.xlabel('Date')
plt.ylabel('Temperature (C)')
plt.title('Monthly Mean Air Temperature')
plt.legend()
plt.grid(True)
plt.show()
```

## The Windspeed

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load CSV
file_path = "hly4935.csv"
df = pd.read_csv(file_path, header=20, low_memory=False)

# Convert 'date' to datetime and set as index
df['date'] = pd.to_datetime(df['date'], errors='coerce')
df.set_index('date', inplace=True)

# Convert 'wdsp' (windspeed) to numeric, coerce errors to NaN
df['wdsp'] = pd.to_numeric(df['wdsp'], errors='coerce')

# Drop rows with missing windspeed
windspeed = df['wdsp']

# Plot Windspeed 
plt.figure(figsize=(12, 5))
plt.plot(windspeed, color='blue')
plt.title("Hourly Windspeed")
plt.xlabel("Date")
plt.ylabel("Windspeed (m/s)")
plt.tight_layout()
plt.show()

```

## The rolling windspeed over 24 hours

```python
# Rolling Windspeed (24 hours) 
rolling_wdsp = windspeed.rolling(window=24).mean() 

plt.figure(figsize=(12, 5))
plt.plot(rolling_wdsp, color='green')
plt.title("24-Hour Rolling Windspeed")
plt.xlabel("Date")
plt.ylabel("Windspeed (m/s)")
plt.tight_layout()
plt.show()
```

## The max windspeed for each day

```python
# Max windspeed for each day 
daily_max_wdsp = windspeed.resample('D').max()

plt.figure(figsize=(12, 5))
plt.bar(daily_max_wdsp.index, daily_max_wdsp, color='orange')
plt.title("Daily Maximum Windspeed")
plt.xlabel("Date")
plt.ylabel("Max Windspeed (m/s)")
plt.tight_layout()
plt.show()
```

## The monthly mean of the daily max windspeeds

```python
# Monthly mean of daily max windspeeds 
monthly_mean_max_wdsp = daily_max_wdsp.resample('M').mean()

plt.figure(figsize=(12, 5))
plt.bar(monthly_mean_max_wdsp.index, monthly_mean_max_wdsp, color='red')
plt.title("Monthly Mean of Daily Maximum Windspeeds")
plt.xlabel("Month")
plt.ylabel("Mean Max Windspeed (m/s)")
plt.tight_layout()
plt.show()
```
